# **LGMVIP May 2023 Data Science Internship task 3**


In this project , I have read an image and converted it into pencil sketch .

Steps involved :

1. Reading the image   
2. Converting the image into gray-scale .  
3. Inverting the gray-scaled image   
4. Blurring the inverted image    
5. Creating the final image by diving the gray scaled imaged by the inverted blurred image .

# **Importing OpenCV2 library**

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

# **Read the image**

```python
input_img = cv2.imread("/content/FRIENDS.jpeg")
cv2_imshow(input_img)
```

# **convert the image to gray scale**

```python
gray_image = cv2.cvtColor(input_img, cv2.COLOR_BGR2GRAY)
cv2_imshow( gray_image)
```

# **Invert the image**

```python
inverted_gray_image = 255 - gray_image
cv2_imshow(inverted_gray_image)
```

#**Blur the image by gaussian blur**

```python
blurred_image = cv2.GaussianBlur(inverted_gray_image, (21, 21), 0)
cv2_imshow(blurred_image)
```

# **Invert the blurred image**

```python
inverted_blurred_image = 255 - blurred_image
cv2_imshow(inverted_blurred_image)
```

# **create the pencil sketch image**

```python
pencil_sketch_image = cv2.divide(gray_image, inverted_blurred_image, scale = 256.0)
cv2_imshow(pencil_sketch_image)
```
